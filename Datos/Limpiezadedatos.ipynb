```python
df_men['weight'] = df_men['weight'].fillna(0) #llenamos de 0 los nulos para luego reemplazarlos por la mediana
```

```python
def convertir_a_kilos(weight):                                    #Tranformamos todo a Kilos 
    value, unit = weight.str.extract('(\d+) (\w+)').values.T
    return pd.to_numeric(value) * (1.0 if unit[0] == 'kg' else 0.45359237)
```

```python
df_men['weight'] = convertir_a_kilos(df_men['weight'])
```

```python
df_men['weight'].value_counts()
```

```python
df_men['weight'] = df_men['weight'].replace(0, np.nan)
```

```python
df_men['weight'] = df_men['weight'].fillna(df_men['weight'].median())
```
